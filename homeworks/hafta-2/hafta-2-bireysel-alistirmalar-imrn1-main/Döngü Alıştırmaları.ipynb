# En uzun Collatz zinciri

```python
N = 10000
dic = {1:1,2:2,3:8}

key = -1
maxL = 2

def isInDic(i):
    if i in dic:
        return True
    return False

def collatz(d,l):
    i=0
    while d[i]!=1: 
        if isInDic(d[i]):
            l+=dic[d[i]]
            break
        else:
            if d[i] %2==0:
                d.append(int(d[i]/2))
            else:
                d.append(d[i]*3+1)
        i+=1
    # print(d)    
    return d,l

for j in range(1,N+1):
    l=0
    if isInDic(j):
        l+=dic[j]
        continue
    d = [j]
    i = 0
    
    d,l = collatz(d,l)

    l +=len(d)-1
    if maxL < l:
        maxL = l
        key = j
    if not isInDic(d[0]):
        dic[d[0]] = l
# print(maxL)
# print(key)
# print(dic)

print(f"1-{N} aralÄ±ÄŸÄ±ndaki:")
print(f"en uzun dizi {maxL} uzunluÄŸunda ve baÅŸlangÄ±Ã§ deÄŸeri {key}")
```

# Ãœstel fonksiyon Taylor serisi

```python
import math

x = float(input("x deÄŸerini girin: "))
tol = 1e-10

t =0
son_terim = 0
n = 0
while True:
    son_terim = x**n / math.factorial(n)
    if son_terim > tol:
        t+=son_terim
    else:
        break    
    n+=1
print("e^x ~",t)
print(n,"terim kullanÄ±ldÄ±")
print("son terim = ",son_terim)
```

```python
import math

N = float(input("KarekÃ¶kÃ¼ alÄ±nacak sayÄ±: "))
x= float(input("Ä°lk tahmin: "))
n = 0
tol=1e-10

while abs(pow(x,2)-N) > tol:
    if abs(pow(x,2)-N) == 0 or abs(pow(x,2)-N) <= tol:
        break
    x = 0.5 * (x + N/x)
    n+=1

print("karekÃ¶k ~",x)

print(n,"iterasyon")
```
